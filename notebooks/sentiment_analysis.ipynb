```python
def visualize_class_distribution(data, title):
    data = data.copy()
    data["label"] = data["label"].apply(lambda x: POL_EMO_INDEXES_TO_CLASSES[x])
    groupped = data.groupby(['label']).count()
    labels = groupped.index.values
    values = groupped.values[:, 0]
    
    fig = plt.figure(figsize=(15, 8))
    
    x_data = np.arange(len(labels))
    
    y_data = values
    
    x = labels
    x_ticks = x_data
    plt.bar(x_data, values, tick_label=labels)
    plt.title(title, fontsize=20)
    plt.ylabel("Number of instances")
    plt.xlabel("label")
    
    xlocs, xlabs = plt.xticks()
    
    for i, v in enumerate(values):
        plt.text(xlocs[i] - 0.10, v + 0.03, str(round(v, 5)), color='black', fontsize=15)
    
    plt.show()
        
visualize_class_distribution(train_data, "all_data")
visualize_class_distribution(train_data.iloc[train_idx,:], "train_data")
visualize_class_distribution(train_data.iloc[test_idx,:], "test_data")
```

# Badania skuteczności predykcji w zależności od zastosowanego taggera i wybranych części mowy

Przeprowadzone badania mają na celu pokazanie jak zastosowany tagger oraz wybrane części mowy mają wpływ na proces klasyfikacji tekstów pod kątem ich wydźwięku. Zbiorem danych na którym przeprowadzono badania jest zbiór POL_EMO składający się z 9324 tekstów oznaczonych jedną z 6 klas symbolizujących jego wydźwięk.

Dla wszystkich badań oraz taggerów zapropoonwany został ten sam, wygenerowany opdowiednio wcześniej podział na zbiór treningowy i testowy w celu zachowania jednorodności podziału dla badanych przypadków.

```python
import time
import pandas as pd

from sklearn.metrics import accuracy_score
from clarin_api_wrapper.krnnt_api_wrapper import KRNNTWrapper
from clarin_api_wrapper.morphodita_wrapper import MorphoditaWrapper
from clarin_api_wrapper.wcrft2_wrapper import Wcrft2Wrapper
from data_loader import load_tagger_preprocesses_json_file, load_polemo_no_segm_file, load_train_test_idx_file
from sentiment_model.sentiment_model import SentimentModel
from settings import PATH_TO_DATA
```

```python
def find_idx_with_tag(data, tags):

    idx = []

    for i in range(0, len(data)):
        found = False
        for token in data[i]:
            if any(tag in token['lex']['ctag'] for tag in tags):
                found = True

        if found:
            idx.append(i)

    return idx


def get_tagger_class(name):
    if name == "krnnt":
        return KRNNTWrapper
    if name == "morphodita":
        return MorphoditaWrapper
    if name == "wcrft2":
        return Wcrft2Wrapper

def do_research():
    taggers = ["krnnt", "morphodita", "wcrft2"]
    parts_of_speech = ["verb", "noun", "adj", None]     # None for all parts of speech
    results = pd.DataFrame(columns=["tagger", "part_of_speech", "accuracy", "time"])
    for tagger in taggers:
        for part_of_speech in parts_of_speech:
            acc, time = do_reasearch_for_one_case(tagger, part_of_speech)
            part_of_speech_str = part_of_speech if part_of_speech is not None else "all"
            row = pd.DataFrame([{"tagger": tagger,
                                 "part_of_speech": part_of_speech,
                                 "accuracy": acc,
                                 "time": time}])
            results = results.append(row, ignore_index=True)
            results.to_csv(f"{PATH_TO_DATA}/results/{tagger}-{part_of_speech_str}.csv")

def do_reasearch_for_one_case(tagger, part_of_speech):
    train_data = load_polemo_no_segm_file("plemo2.0-no-segm.txt")
    preprocessed_data = load_tagger_preprocesses_json_file(
        f"{PATH_TO_DATA}/taggers_output/{tagger}/plemo_no_segm_data.json")
    train_idx, test_idx = load_train_test_idx_file(f"{PATH_TO_DATA}/train_test_idx.pkl")
    x_data, y_data = train_data.values[:, 0], train_data.values[:, 1].astype(int)

    x_train, y_train = preprocessed_data[train_idx], y_data[train_idx]
    x_test, y_test = preprocessed_data[test_idx], y_data[test_idx]

    preprocessed_train, preprocessed_test = preprocessed_data[train_idx], preprocessed_data[test_idx]

    model = SentimentModel(tagger_wrapper=Wcrft2Wrapper())

    start = time.time()
    model.fit(preprocessed_train, y_train, part_of_speech=part_of_speech, tagger_preprocessed=True)
    elapsed = time.time() - start
    y_pred = model.predict(preprocessed_test, part_of_speech=part_of_speech, tagger_preprocessed=True)

    acc = accuracy_score(y_test, y_pred)
    part_of_speech_str = part_of_speech if part_of_speech is not None else "all"
    print("Tagger: {} Part of speech: {} Accuracy: {}".format(tagger,
                                                              part_of_speech_str,
                                                              acc))
    return acc, elapsed
```

```python
do_research()
```

```python
def compare_taggers(data, title):
    results = data.values
    labels = results[:, 0]
    accuracy = results[:, 2]
    
    fig = plt.figure(figsize=(15, 8))
    
    x_data = np.arange(len(labels))
    
    y_data = accuracy
    
    x = labels
    x_ticks = x_data
    plt.bar(x_data, accuracy, tick_label=labels)
    plt.title(title, fontsize=20)
    plt.ylabel("Accuracy")
    plt.xlabel("Taggers")
    
    xlocs, xlabs = plt.xticks()
    
    for i, v in enumerate(accuracy):
        plt.text(xlocs[i] - 0.10, v - 0.04, str(round(v, 5)), color='white', fontsize=15)
    
    plt.show()


def compare_part_of_speeches(data, title):
    results = data.values
    labels = results[:, 1]
    accuracy = results[:, 2]
    fig = plt.figure(figsize=(15, 8))
    
    x_data = np.arange(len(labels))
    
    y_data = accuracy
    
    x = labels
    x_ticks = x_data
    plt.bar(x_data, accuracy, tick_label=labels)
    plt.title(title, fontsize=20)
    plt.ylabel("Accuracy")
    plt.xlabel("Taggers")
    
    xlocs, xlabs = plt.xticks()
    
    for i, v in enumerate(accuracy):
        plt.text(xlocs[i] - 0.10, v - 0.04, str(round(v, 5)), color='white', fontsize=15)
    
    plt.show()
    



results = pd.read_csv(f"{PATH_TO_DATA}/results/wcrft2-all.csv", index_col=0).fillna("all")    
```

```python
compare_part_of_speeches(results[results['tagger'] == "krnnt"], "Tagger - krnnt")
compare_part_of_speeches(results[results['tagger'] == "morphodita"], "Tagger - morphodita")
compare_part_of_speeches(results[results['tagger'] == "wcrft2"], "Tagger - wcrft2")
```

Na podstawie wyników przeprowadzonych badań można zauważyć, że najlepsze wyniki klasyfikator, dla każdego z zastosowanych taggerów osiąga dokonując predykcji na podstawie wszystkich słów znajdujących się w klasyfikowanym zdaniu.

Co ciekawe, każdy tagger osiąga podobne wyniki w zależności od zastosowanych części mowy. Dla pojedynczych części mowy najlepsze są przymiotniki, a najgorzej spisują się czasowniki. Intuicyjnie ma to sens, ze względu na funkcję jaką przymiotniki pełnią w języku.

```python
compare_taggers(results[results['part_of_speech'] == "all"], "Part of speech - ALL")
compare_taggers(results[results['part_of_speech'] == "verb"], "Part of speech - Verbs")
compare_taggers(results[results['part_of_speech'] == "noun"], "Part of speech - Nouns")
compare_taggers(results[results['part_of_speech'] == "adj"], "Part of speech - Adj")
```
